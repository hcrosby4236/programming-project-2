```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("james_bond_data_cleansed.csv")
print(data)
```

```python
plt.scatter(data['imdb'], data['rotten_tomatoes'])
plt.xlabel('IMDB Rating')
plt.ylabel('Rotten Tomatoes Rating')
plt.title('Scatter Plot of Ratings')
plt.show()
# I would say that this plot has a moderate postive correlation
```

```python
plt.scatter(data['imdb'], data['bond_kills'])
plt.xlabel('IMDB Ratings')
plt.ylabel('Bond Kills')
plt.title('IMDB Ratings v. Bond Kills')
plt.show()
# There is no correlation here
```

```python
plt.plot(data["release_year"],data["income_world"],'-o', label='world income',color='red')
plt.plot(data['release_year'], data['movie_budget'],'b-o',label='movie budget')
plt.legend()

plt.show()
# Everything has been uphill since about 2011
```

```python
world_income = data['income_world']
budget = data['movie_budget']
ratio = (world_income / budget)
plt.bar(data['release_year'],ratio)
plt.xlabel('Release Year')
plt.ylabel('World Income/Budget Ratio')
plt.title('World Income-Budget Ratio over the Years')
plt.show()
# I think that over time they began to lose alot of money making these movies. 
```
